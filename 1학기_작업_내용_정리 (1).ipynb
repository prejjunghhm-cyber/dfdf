최근접 점쌍 구하기

```python
import math

def distance(p1, p2):
  return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)


def closest_pair(points):
  n = len(points)

  if n == 2:
      return distance(points[0], points[1])

  if n == 3:
    d1 = distance(points[0], points[1])
    d2 = distance(points[0], points[2])
    d3 = distance(points[1], points[2])
    return min(d1, d2, d3)

  mid = n // 2
  mid_x = points[mid][0]

  left_points = points[:mid]
  right_points = points[mid:]

  left_min = closest_pair(left_points)
  right_min = closest_pair(right_points)

  d = min(left_min, right_min)

  candidates = []

  for point in points:
    if abs(point[0] - mid_x) < d:
      candidates.append(point)

  candidates.sort(key=lambda x: x[1])

  for i in range(len(candidates)):
    j = i + 1

    while j < len(candidates) and candidates[j][1] - candidates[i][1] < d:
      d = min(d, distance(candidates[i], candidates[j]))
      j += 1

  return d
```

수식과 괄호 삽입

```python
def calculate(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b


def divide_and_conquer(expression):

  #정복단계
    if expression.isdigit():
        return [int(expression)]

    results = []

   #분할단계
    for i in range(len(expression)):
        if expression[i] in '+-*':
            op = expression[i]

            left_expr = expression[:i]
            right_expr = expression[i + 1:]

            left_results = divide_and_conquer(left_expr)
            right_results = divide_and_conquer(right_expr)

     #결합단계
            for left in left_results:
                for right in right_results:
                    results.append(calculate(left, right, op))

    return results


expression = input().strip()

answer = sorted(set(divide_and_conquer(expression)))

print(answer)
```

숫자 골라 최대합 구하기

```python
def main():
    n, m, k = map(int, input().split())

    if not (1 <= n <= 10 and 1 <= m <= 10):
        print("n과 m은 1 이상 10 이하의 정수여야 합니다.")
        return

    if not (1 <= k <= min(n * m, 4)):
        print("k 값이 조건을 벗어났습니다.")
        print("k는 1 이상이고, n*m과 4 중 최솟값 이하여야 합니다.")
        return

    max_pick = (n * m + 1) // 2
    if k > max_pick:
        print("k개를 서로 인접하지 않게 고를 수 없습니다.")
        return

    board = []
    for _ in range(n):
        row = list(map(int, input().split()))

        if len(row) != m:
            print(f"각 행에는 숫자가 {m}개씩 들어와야 합니다.")
            return

        board.append(row)

    visited = [[False] * m for _ in range(n)]

    answer = -10**18

    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]

    def can_pick(x, y):
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0 <= nx < n and 0 <= ny < m:
                if visited[nx][ny]:
                    return False

        return True

    def backtracking(index, count, total):
        nonlocal answer

        if count == k:
            answer = max(answer, total)
            return

        if index == n * m:
            return

        if count + (n * m - index) < k:
            return

        x = index // m
        y = index % m

        if can_pick(x, y):
            visited[x][y] = True
            backtracking(index + 1, count + 1, total + board[x][y])
            visited[x][y] = False

        backtracking(index + 1, count, total)

    backtracking(0, 0, 0)

    print(answer)


main()
```

k개 제거 후 최댓값 구하기

```python
def biggest_number(number, k):
    stack = []

    for digit in number:
        while stack and k > 0 and stack[-1] < digit:
            stack.pop()
            k -= 1

        stack.append(digit)
        print("현재 숫자:", digit, "스택 상태:", stack)

    if k > 0:
        stack = stack[:-k]

    return ''.join(stack)
```
